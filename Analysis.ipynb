```python
import pandas as pd
import sqlalchemy
from sqlalchemy.types import Integer, Text, String, Float
import psycopg2
import matplotlib.pyplot as plt
import numpy as np
from config import user, pas
```

```python
#connect to the psql database
engine = sqlalchemy.create_engine(f'postgresql+psycopg2://{user}:{pas}@127.0.0.1:5432/london_trips')
```

```python
#Connecting the new tables from the psql database
reason_length_il = pd.read_sql_table("reason_length_il",con=engine)
reason_length_ll = pd.read_sql_table("reason_length_ll",con=engine)
pop_travel_il = pd.read_sql_table("pop_travel_il",con=engine)
pop_travel_ll = pd.read_sql_table("pop_travel_ll",con=engine)
```

```python
#plot the international VS Reason for travel & save the plot
x_axis = np.arange(len(reason_length_il))
plt.figure()
bar = reason_length_il.plot.bar(x='reason_for_travel')
plt.ylabel("trip length count")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ['VFR: 4-7 nights', 'VFR: 8-14 nights', 'VFR: 1-3 nights', 'VFR: 15+ nights',
                            'Study: 4-7 nights', 'Study: 8-14 nights', 'Study: 1-3 nights', 'Study: 15+ nights',
                            'Miscellaneous: 1-3 nights', 'Miscellaneous: 8-14 nights', 'Miscellaneous: 4-7 nights', 'Miscellaneous: 15+ nights',
                            'Holiday: 8-14 nights', 'Holiday: 1-3 nights', 'Holiday: 15+ nights', 'Holiday: 4-7 nights',
                            'Business: 1-3 nights', 'Business: 8-14 nights', 'Business: 15+ nights', 'Business: 4-7 nights'],rotation=90)    
plt.savefig("images/reason_length_il.png",bbox_inches='tight')
```

```python
#plot the local VS Reason for travel & save the plot
x_axis = np.arange(len(reason_length_ll))
plt.figure()
bar = reason_length_ll.plot.bar(x='reason_for_travel')
plt.ylabel("trip length count")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Other: 0.0",
                            "Member portfolio/Council business (Elected Mem...: 2.0",
                            "Member portfolio/Council business (Elected Mem...: NaN",
                            "Member portfolio/Council business (Elected Mem...: 0.0",
                            "Member portfolio/Council business (Elected Mem...: 1.0",
                            "Meeting with other public sector organisations: 0.0",
                            "Meeting with other public sector organisations: 1.0",
                            "Meeting with other external bodies: 0.0",
                            "LCC hosted event  Externally funded: 0.0",
                            "Conference - as an exhibitor/presenter: 0.0",
                            "Conference - as an exhibitor/presenter: .0",
                            "Conference - as an attendee: 1.0",
                            "Conference - as an attendee: 0.0",
                            "Conference - as an attendee: NaN"],rotation=90)        
plt.savefig("images/reason_length_ll.png",bbox_inches='tight')
```

```python
#plot the travel type & save the plot
x_axis = np.arange(len(pop_travel_il))
plt.figure()
bar = pop_travel_il.plot.bar(x='travel_type')
plt.ylabel("trip length count")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Tunnel",
                            "Sea",
                            "Air"],rotation=90)        
plt.savefig("images/pop_travel_il.png",bbox_inches='tight')
```

```python
#plot the travel type & save the plot
x_axis = np.arange(len(pop_travel_ll))
plt.figure()
bar = pop_travel_ll.plot.bar(x='travel_type')
plt.ylabel("trip length count")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Train",
                            "Flight"],rotation=90)        
plt.savefig("images/pop_travel_ll.png",bbox_inches='tight')
```
